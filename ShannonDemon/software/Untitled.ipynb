```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import itertools
import seaborn as sns
from simulator import tom_return_given_trajectory, steve_return_given_trajectory, ben_return_given_trajectory
```

```python
TIME = 100
plot = {'steve': [], 'ben': [], 'tom': []}

num_samples = 300
def create_random_traj(time):
    return [int(np.random.random() <= 0.5) for i in range(time)]

trajectories = [create_random_traj(TIME) for i in range(num_samples)]
for idx, traj in enumerate(trajectories):
    rewards, _ = steve_return_given_trajectory(traj)
    plot['steve'].append(rewards)

    rewards, _ = ben_return_given_trajectory(traj)
    plot['ben'].append(rewards)

    
    rewards, _ = tom_return_given_trajectory(traj)
    plot['tom'].append(rewards)
    
```

```python
sns.set(rc={"figure.figsize": (15, 15)})
df = pd.DataFrame(plot['steve'])
ax = df.where(df.ge(0)).T.plot(color='green', legend=False, fontsize=18)
df.where(df.lt(1)).T.plot(ax=ax, color='red', legend=False)
# ax = df.T.plot(figsize=(20, 20), legend=False)
ax.set_yscale('log')
ax.set_ylabel('Take home cash', fontsize=18)
ax.set_xlabel('Number of days', fontsize=18)
ax.set_title("Steve's strategy of investing", fontsize=20)
```

```python
sns.set(rc={"figure.figsize": (15, 15)})
df = pd.DataFrame(plot['ben'])
ax = df.where(df.ge(0)).T.plot(color='green', legend=False, fontsize=18)
df.where(df.lt(1)).T.plot(ax=ax, color='red', legend=False)
# ax = df.T.plot(figsize=(20, 20), legend=False)
# ax.set_yscale('log')
ax.set_ylabel('Take home cash', fontsize=18)
ax.set_xlabel('Number of days', fontsize=18)
ax.set_title("Ben's strategy of investing", fontsize=20)
```

```python
sns.set(rc={"figure.figsize": (15, 15)})
df = pd.DataFrame(plot['tom'])
ax = df.where(df.ge(0)).T.plot(color='green', legend=False, fontsize=18)
df.where(df.lt(1)).T.plot(ax=ax, color='red', legend=False)
# ax = df.T.plot(figsize=(20, 20), legend=False)
ax.set_yscale('log')
ax.set_ylabel('Take home cash', fontsize=18)
ax.set_xlabel('Number of days', fontsize=18)
ax.set_title("Tom's strategy of investing", fontsize=20)
```
